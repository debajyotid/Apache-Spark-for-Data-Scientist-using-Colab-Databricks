```
#In this notebook, we talk about Koalas, which is a distributed dataframe API that can run on top of Spark. Today, Python is the de-facto language for many data & analytics workload. Python supports rich gamut of libraries like numpy, andas, scipy, etc. which makes the different taks of machine learning, data anaylsis and statistical analysis pretty easy. However, Pandas is extremely memory dependent. It can' be run in a distributed fashion & depends on a single node system to load the data into memory & process it. Thus it is not vey ideal for large datasets. While it is possible to chunk the data, it provides a limitation on data processing. 
#This is where distributed framework APIs like DASK came into picture which runs Pandas in a more distributed way. Then we have cuDF from RAPIDS, which leverages GPU and has a resemblance to Pandas. 
#Spark is a very efficient processing engine for running large datasets. However it comes with a huge learning curve for someone comfortable working on Pandas. How to make it simpler is where Koalas comes into play. The idea of Koalas framework resembles Pandas API as it is.It makes the learning curve between Pandas and Spark much less steeper. It runs on top of PySpark. In addition, it also helps transition old projects written in Pandas to PySpark. Lastly, as a major divergence from Spark, Koalas is not a lazy executioner like Spark. So it's similar to Pandas once again in this case
```

```
ks.set_option('display.max_rows',50)
#w/o this parameter, koalas will display all the rows available in a dataframe
```

```
loan_data=ks.read_csv('dbfs:/FileStore/tables/LoanStats_2018Q4.csv')
#shifting to Koalas dataframe (syntax is same as pd.read_csv & leveraging the distributed file system, rather than local file system)
```

```
loan_data['loan_status'].value_counts()
```

```
loan_data.head()
```

```
ld_sel=loan_data[['loan_amnt','grade','term','loan_status']]
```

```
ld_sel['loan_status'].value_counts()
```

```
ld_sel.groupby('grade').agg({'grade':'count'})
```

```
ld_sel.groupby('grade').agg({'grade':'count'})/ld_sel['grade'].count()
```

```
#Koalas disallows the operations on different DataFrames (or Series) by default to prevent expensive operations. It internally performs a join operation which can be expensive in general.
#This can be enabled by setting compute.ops_on_diff_frames to True to allow such cases

ks.set_option('compute.ops_on_diff_frames', True)
ld_tmp=ks.concat([ld_sel.groupby('grade').agg(grade_count=('grade','count')),
                  ld_sel.groupby('grade').agg(grade_pct=('grade','count'))/ld_sel['grade'].count()],axis=1)
```

```
ld_tmp.head(8)
```

```
ld_tmp.reset_index()
```

```
ks.sql("SELECT * FROM {ld_tmp} WHERE grade_count > 300") #this is an added advantage of Koalas
```

```
ld_tmp.sort_index()
```

```
ld_tmp.sort_values(by='grade_pct')
```

```
import numpy as np
#Koalas allows use of other Python libraries like Numpy also
ld_tmp.apply(np.sum)
```

```
ld_tmp=ld_tmp.dropna()
```

```
display(ld_tmp)
```

```
ld_tmp.apply(lambda x:x*2)
```

```
def multiply(x)->ks.Series[np.float64]:
  return x*2
```

```
ld_tmp.apply(multiply)
```

```
display(ld_tmp.plot.bar())
#Another advantage of Koalas over Spark. We can leverage the built-in plotting function
#Earlier with Spark, we had to get the data to the local node using to_pandas() and then use matplotib/seaborn. Koalas provides that support out of the box
```
